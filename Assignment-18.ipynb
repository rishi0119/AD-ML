```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
df = pd.read_csv(r"C:\Users\madha\Downloads\Day_18_Tours_and_Travels.csv")
display(df.info())
display(df.head())
df['Customer_Age'].fillna(df['Customer_Age'].median(), inplace=True)
df['Rating'].fillna(df['Rating'].median(), inplace=True)
df['Review_Text'].fillna("No Review", inplace=True)
df.drop_duplicates(subset=['Review_Text', 'Customer_Age', 'Rating'], keep='first', inplace=True)
df['Rating'] = df['Rating'].clip(1, 5)
df['Tour_Package'] = df['Tour_Package'].str.title()
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Package_Price'])
plt.show()
df['Package_Price'] = df['Package_Price'].clip(lower=df['Package_Price'].quantile(0.05), 
                                               upper=df['Package_Price'].quantile(0.95))
encoder = LabelEncoder()
df['Tour_Package'] = encoder.fit_transform(df['Tour_Package'])
scaler = MinMaxScaler()
df[['Customer_Age', 'Package_Price']] = scaler.fit_transform(df[['Customer_Age', 'Package_Price']])
df.to_csv("cleaned_travel_reviews.csv", index=False)
display(df.head())
```
