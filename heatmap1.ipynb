```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset('tips')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
df['smoker']=lb.fit_transform(df['smoker'])
df['sex']=lb.fit_transform(df['sex'])
df['time']=lb.fit_transform(df['time'])
df['day']=lb.fit_transform(df['day'])
```

```python
df.dtypes
```

```python
df.head(3)
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
sns.heatmap(np.abs(df.corr()),cmap='Purples',annot=True)
plt.show()
```

```python
sns.heatmap(np.abs(df.corr()),cmap='Oranges',annot=True)
plt.show()
```

```python
df.describe()
```

```python
"""
scaling techniques
1) standardization ------> mean 0 and std of 1
2) normalization   ------> range of 0
"""
from sklearn .preprocessing import StandardScaler
sc=StandardScaler()
sc
```

```python
sc_array=sc.fit_transform(df)
print(sc_array)
sc_array.shape
```

```python
sc_df=pd.DataFrame(sc_array,columns=df.columns)
sc_df.head(3)
```

```python
sc_df.describe()
```

```python
sc_df.corr()
```
