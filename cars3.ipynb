```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cars[['HP','MPG','VOL','SP','WT']] = scaler.fit_transform(cars[['HP','MPG','VOL','SP','WT']])
```

```python
np.round(cars.describe(),3)
```

```python
import statsmodels.formula.api as smf
m1_hp=smf.ols("MPG~HP",data=cars).fit()
m1_hp.rsquared
m1_v=smf.ols("MPG~VOL",data=cars).fit()
m1_v.rsquared
```

```python
m1_v.params
```

```python
m1_sp=smf.ols("MPG~SP",data=cars).fit()
m1_sp.rsquared
```

```python
m1_wt=smf.ols("MPG~WT",data=cars).fit()
m1_wt.rsquared
```

```python
model=smf.ols("MPG~HP+VOL+SP+WT",data=cars).fit()
model.rsquared
```

```python
model.params
```
